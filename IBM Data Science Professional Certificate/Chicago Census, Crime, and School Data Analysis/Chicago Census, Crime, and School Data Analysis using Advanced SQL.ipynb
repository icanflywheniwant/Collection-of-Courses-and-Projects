### Advanced SQL for Data Engineers: Problem 1 (Exercise 1)

##### Write and execute a SQL query to list the school names, community names and average attendance for communities with a hardship index of 98.



```python
%%sql 
Select cp.NAME_OF_SCHOOL,cp.COMMUNITY_AREA_NUMBER,cp.AVERAGE_STUDENT_ATTENDANCE,cp.average_teacher_attendance
from CHICAGO_PUBLIC_SCHOOLS as Cp LEFT JOIN CENSUS_DATA as cd On cp. COMMUNITY_AREA_NUMBER = cd. COMMUNITY_AREA_NUMBER AND
cd. hardship_index = 98;
```

### Advanced SQL for Data Engineers: Problem 2 (Exercise 1)

##### Write and execute a SQL query to list all crimes that took place at a school. Include case number, crime type and community name.



```python
%%sql 
Select cp.CASE_NUMBER, cd.COMMUNITY_AREA_NAME, cp.PRIMARY_TYPE
from CHICAGO_CRIME_DATA as cp LEFT OUTER JOIN CENSUS_DATA as cd On cp.COMMUNITY_AREA_NUMBER = cd. COMMUNITY_AREA_NUMBER
Where cp. LOCATION_DESCRIPTION LIKE '%SCHOOL%';
```
