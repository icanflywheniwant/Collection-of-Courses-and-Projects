- Welcome back! On this task, we are going to learn how we merge and concatenate datasets and variables on Xarray. This is important if you are generating data or handling several sources, as it allows you to mix everything together into a single object.
- So let's start by opening the task 6 notebook. Run everything as it is.

```python
import xarray as xr
```

```python
ds = xr.open_dataset('data.nc').sel(expver=1)
```

- Now suppose that we create a new variable, which is simple the year of the points that we have. We can get that through da equals ds time dt year, da. Run it.

```python
da = ds.time.dt.year
da
```

- Notice that the year variable has only the time dimension. Now the question is: how you can plug that separate array into a dataset? This is done through the merge method.
- To see it happening, open a new block and type with me: ds 1 equals ds merge da ds 1. Run it.

```python
ds_1 = ds.merge(da)
ds_1
```

- Notice that now we have the year variable, and that it does have only the time dimension.
- Another variation is to take the sum of the squares of the latitude and the longitude, however on this one we will be needing to set the variable name as Xarray will not be able to infer about it.
- To see what I mean, open a new cell, and type with me: da equals ds lontitude squared plus ds latitude squared, da name equals squared distance, ds 2 equals ds merge da. ds 2. Run it.

```python
da = ds.longitude ** 2 + ds.latitude ** 2
da.name = 'squared_distance'
ds_2 = ds.merge(da)
ds_2
```

- Now we have a variable that has the longitude an latitude coordinates!
- In order to merge them, we will be using of course the merge method. Open a new block, and type with me: ds 3 equals ds 1 merge ds2. ds3. Run it.

```python
ds_3 = ds_1.merge(ds_2)
ds_3
```

- Now we have everyone together! Another interesting feature to get to know is dataset concatenation, which allows you to create new dimensions while appending data.
- To see what I mean, open a new block and let's create two new datasets. The first one, ds square root is equals to ds 3 power by 1 per 2. The second, ds cube root, is equal to ds 3 power 1 per 3. We are getting all the variables, and getting their square and cube roots.
- Now let's create a new dataset which contains the information for the original and the new datasets. IN order to do so, we write ds 4 equals xr concat ds 3 ds square root ds cube root new dim. Run it.

```python
ds_square_root = ds_3 ** (1 / 2)
ds_cube_root = ds_3 ** (1 / 3)
ds_4 = xr.concat([ds_3, ds_square_root, ds_cube_root], 'new_dim')
ds_4
```

- What we are doing here, is to create a new dimension, which is the new dim. This dimension keeps track of the square and cube root datasets that we've created, and you can see at the top that we have three values for new dim, and that all our variables now also have a dimension for it.
- To see how it manifests, let's open a new block and first plot the total precipation. ds sel time 2019 01 01 tp plot. This is the total precipation for the original dataset, and run it.

```python
ds.sel(time='2019-01-01').tp.plot()
```

- Fine, now let's do the same but for the new dataset and by selecting the first value of the new dim. Copy and paste into a new block, exchange ds by ds4, and put a new dim equals zero argument. Run it.

```python
ds_4.sel(new_dim=0, time='2019-01-01').tp.plot()
```

- Same thing, now let's copy and paste so that we see if we select new dim equals 1. Run it.

```python
ds_4.sel(new_dim=1, time='2019-01-01').tp.plot()
```

- Now it is different? What about the last one? Copy and paste and put new dim equals two. Run

```python
ds_4.sel(new_dim=2, time='2019-01-01').tp.plot()
```

- It changed again! The concatenation worked! Feel free to pause the video to see how things are when you change the dates.
